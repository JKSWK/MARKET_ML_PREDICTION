```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import pandas as pd
```

```python
## BADANIE CAŁEJ POPULACJI WZGLĘDEM WIG-20
```

```python
X = df.drop("WIG20_d", axis=1)
y = df.WIG20_d
```

```python
display(X)
display(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=42, test_size=0.2)
```

```python
X.describe()
```

```python
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)
```

```python
reg_model = MLPRegressor(
    hidden_layer_sizes=(32, 16, 8, 4),
    activation='relu',
    solver='adam',
    random_state=42,
    max_iter=1000,
    batch_size=64)

reg_model.fit(X_train_sc, y_train)
```

```python
y_pred = reg_model.predict(X_test_sc)
y_pred_train = reg_model.predict(X_train_sc)
print(f"R2 score: {r2_score(y_pred, y_test)}")
print(f"R2 score train : {r2_score(y_pred_train, y_train)}")
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print(f"MAPE : {mean_absolute_percentage_error(y_pred, y_test)}")
print(f"MAPE  train : {mean_absolute_percentage_error(y_pred_train, y_train)}")
```
